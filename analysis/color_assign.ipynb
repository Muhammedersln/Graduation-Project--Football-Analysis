```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
image_path = "../output_videos/cropped_image.jpg"
image = cv2.imread(image_path)
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

oyuncuyu yarıya bölme

```python
top_half_image = image[0: int(image.shape[0]/2),:]
plt.imshow(top_half_image)
plt.show()
```

cluster the image


```python
image_2d = top_half_image.reshape(-1,3)

kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(image_2d)

#get cluster labels
labels = kmeans.labels_

#reshape the into the original image
cluster_image = labels.reshape(top_half_image.shape[0], top_half_image.shape[1])

#display
plt.imshow(cluster_image)
plt.show
```

```python
corner_cluster = [cluster_image[0,0], cluster_image[0,-1], cluster_image[-1,0], cluster_image[-1,-1]]
non_player_cluster = max(set(corner_cluster), key=corner_cluster.count)
print(non_player_cluster)
```

```python
player_cluster = 1-non_player_cluster
print(player_cluster)
```

```python
kmeans.cluster_centers_[player_cluster]
```
